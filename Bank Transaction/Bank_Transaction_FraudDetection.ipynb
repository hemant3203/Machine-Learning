```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['TransactionType', 'Location', 'Channel', 'CustomerOccupation']

label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df.columns
```

```python
df.drop(['TransactionID', 'AccountID', 'DeviceID', 'IP Address', 'PreviousTransactionDate', 'TransactionDate', 'MerchantID'], axis=1, inplace = True)
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pca = PCA(n_components=2)
reduced_df = pca.fit_transform(df)
df['PCA1'] = reduced_df[:, 0]
df['PCA2'] = reduced_df[:, 1]
```

```python
kmeans = KMeans(n_clusters=5)
df['Cluster'] = kmeans.fit_predict(df[['PCA1', 'PCA2']])
```

```python
df['ClusterDistance'] = np.sqrt((df['PCA1'] - kmeans.cluster_centers_[df['Cluster']][:, 0])**2 + (df['PCA2'] - kmeans.cluster_centers_[df['Cluster']][:, 1])**2)
```

```python
threshold = df['ClusterDistance'].quantile(0.95)
df['Anomaly'] = df['ClusterDistance'] > threshold
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='viridis', style=df['Anomaly'])
plt.title('PCA Clusters with Anomaly Detection')
plt.legend(title='Cluster')
plt.show()
```
